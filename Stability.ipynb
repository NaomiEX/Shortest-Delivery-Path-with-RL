## Evolutionary Stability

From the game theory perspective, we have the following payoff matrix.

![evolutionary](img/evolution_stationary.png)

Let's transform it into the evolutionary perspective. Let p be the propotion of probability choosing the coop strategy, blue dots be the payoff for coop strategy and red dot be the payoff for defect strategy. Using this, we can have the following graph.

If 1\% of the population suddenly changes to defect in a world filled with coop, p = 0.99. We note that following the coop strategy would yield a much higher payoff than being defect. Hence, the agents would stay to coop and those following defect would slowly disappear. This shows that the coop strategy is evolutionarily stable.

On the other hand, if 1\% of the population suddenly changes to defect in a world filled with defect, p = 0.01. We notice that the defect strategy yields better payoff and hence the agents would stick to defect strategy. This is yet another equilibrium state.

Both coop and defect strategies are resistant to alternative strategy according to the graph. When $\frac{1}{3}$ of the population follows coop strategy, on average, the payoff for both groups would be the same. With that, at state $p = \frac{1}{3}$, it is unchanged and become an equilibrium.

```python
from matplotlib import pyplot as plt

def plot_stability_graph(coop_payoff, defect_payoff):
    fig, ax = plt.subplots()
    
    # NOTE: Assume first element in payoff is at p = 0 while second element in payoff is at p = 1
    y = [*coop_payoff, *defect_payoff]
    x = [0, 1, 0, 1]
    c = ['lightblue', 'darkblue', 'lightcoral', 'deeppink']
    
    ax.scatter(x, y, c=c, s=20, zorder=100)
    ax.plot([0, 1], coop_payoff, c='blue', label='Coop')
    ax.plot([0, 1], defect_payoff, c='orange', label='Defect')
    ax.axvline([1/3], ymin=0, ymax=0.35, linestyle='--')
    ax.set_xlabel('p')
    ax.set_ylabel('Reward')
    ax.legend()
    plt.show()
    
```

```python
plot_stability_graph((0, 60), (20, 20))
```
